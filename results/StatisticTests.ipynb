Testar normalidade dos dados:

```python
import pandas as pd
from scipy.stats import shapiro

dados = pd.read_csv('results_Modified.csv')

# Extrair a coluna 'M_pop_20'
Modified_m_pop_20 = dados['M_pop_20']

# Realizar o teste de Shapiro-Wilk
stat, p_valor = shapiro(Modified_m_pop_20)

# Imprimir os resultados
print(f'Estatística do teste: {stat}')
print(f'P-valor: {p_valor}')

# Interpretar os resultados
alpha = 0.05
if p_valor > alpha:
    print('Os dados parecem seguir uma distribuição normal (não podemos rejeitar a hipótese nula)')
else:
    print('Os dados não seguem uma distribuição normal (rejeitamos a hipótese nula)')
```

```python
import pandas as pd
from scipy.stats import shapiro

dados = pd.read_csv('results_original.csv')

# Extrair a coluna 'M_pop_20'
Original_m_pop_20 = dados['min_max_x_op_pmx_mut_swap_pop_20']

# Realizar o teste de Shapiro-Wilk
stat, p_valor = shapiro(Original_m_pop_20)

# Imprimir os resultados
print(f'Estatística do teste: {stat}')
print(f'P-valor: {p_valor}')

# Interpretar os resultados
alpha = 0.05
if p_valor > alpha:
    print('Os dados parecem seguir uma distribuição normal (não podemos rejeitar a hipótese nula)')
else:
    print('Os dados não seguem uma distribuição normal (rejeitamos a hipótese nula)')
```

```python
from scipy.stats import mannwhitneyu

# dados_normal e dados_nao_normal representam os dois conjuntos de dados
estatistica, p_valor = mannwhitneyu(Modified_m_pop_20, Original_m_pop_20)

# Imprimir os resultados
print(f'Estatística do teste U de Mann-Whitney: {estatistica}')
print(f'P-valor: {p_valor}')

# Interpretar os resultados
alpha = 0.05
if p_valor > alpha:
    print('Não há evidências suficientes para rejeitar a hipótese nula de igualdade de distribuições.')
else:
    print('Há evidências para rejeitar a hipótese nula de igualdade de distribuições.')
```

```python
from scipy.stats import ttest_ind

# dados_normal e dados_nao_normal representam os dois conjuntos de dados
estatistica, p_valor = ttest_ind(Modified_m_pop_20, Original_m_pop_20, equal_var=False)

# Imprimir os resultados
print(f'Estatística do teste t de Welch: {estatistica}')
print(f'P-valor: {p_valor}')

# Interpretar os resultados
alpha = 0.05
if p_valor > alpha:
    print('Não tem evidências para rejeitar a hipótese nula.')
else:
    print('Rejeita a hipótese nula')
```

```python
from scipy.stats import ttest_ind

# Modified_m_pop_20 e Original_m_pop_20 representam os dois conjuntos de dados
estatistica, p_valor = ttest_ind(Modified_m_pop_20, Original_m_pop_20)

# Imprimir os resultados
print(f'Estatística do teste t de Student: {estatistica}')
print(f'P-valor: {p_valor}')

# Interpretar os resultados
alpha = 0.05
if p_valor > alpha:
    print('Não há evidências suficientes para rejeitar a hipótese nula de igualdade de médias.')
else:
    print('Há evidências para rejeitar a hipótese nula de igualdade de médias.')
```
