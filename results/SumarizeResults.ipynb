```python
import os
import json
import pandas as pd

def obter_ultimo_best_fitness(caminho_pasta_execucao, tipo):
    caminho_arquivo_json = os.path.join(caminho_pasta_execucao, f"{tipo}_statistics.json")

    if os.path.exists(caminho_arquivo_json):
        with open(caminho_arquivo_json, 'r') as arquivo_json:
            dados_json = json.load(arquivo_json)

            # Verificar se "best_fitness" existe e não está vazio
            if "best_fitness" in dados_json and dados_json["best_fitness"]:
                # Obter o último valor de "best_fitness"
                return dados_json["best_fitness"][-1]

    return None

def obter_resultados_pasta(caminho_pasta_raiz, tipos):
    resultados = {}

    for tipo in tipos:
        resultados[tipo] = []
        for i in range(30):  # Para cada pasta de execution_0 a execution_29
            nome_pasta_execucao = f"execution_{i}"
            caminho_pasta_execucao = os.path.join(caminho_pasta_raiz, tipo, nome_pasta_execucao)

            ultimo_best_fitness = obter_ultimo_best_fitness(caminho_pasta_execucao, "minmax")
            resultados[tipo].append(ultimo_best_fitness)

    return resultados

# Substitua os caminhos para os diretórios raiz conforme necessário
caminho_pasta_raiz = "C:/Users/User/Downloads/results-original/original_min_max"

tipos_operadoresS = [
    "min_max_x_op_order_mut_scramble_pop_20",
    "min_max_x_op_order_mut_swap_pop_20",
    "min_max_x_op_pmx_mut_scramble_pop_20",
    "min_max_x_op_pmx_mut_swap_pop_20"
]

# Iterar sobre os diferentes tipos de operadores
resultados_dict = obter_resultados_pasta(caminho_pasta_raiz, tipos_operadoresS)

# Criar DataFrame do pandas com colunas para os resultados
df = pd.DataFrame(resultados_dict)

# Salvar o DataFrame em um arquivo CSV ou outro formato desejado
df.to_csv('results_Original_MinMax.csv', sep=',', index=False)
```
